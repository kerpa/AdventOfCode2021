```python
import numpy as np
```

```python
data_radar = np.genfromtxt('input_radar.csv', delimiter=',', dtype="U75")

data_radar
```

```python
data_radar_list = data_radar.tolist()
data_radar_list
```

```python
#split each string into 2 strings
data_radar_list_split = [x.split(' ') for x in data_radar_list]
data_radar_list_split
```

```python
depth , horizontal_position, aim  = 0, 0, 0
```

```python
def update_position(sub_list):
    global depth, horizontal_position, aim
    if sub_list[0] == 'down':
        aim += int(sub_list[1])
    elif sub_list[0] == 'up':
        aim -= int(sub_list[1])
    elif sub_list[0] == 'forward':
        horizontal_position += int(sub_list[1])
        depth += aim * int(sub_list[1])
    return depth, horizontal_position, aim
```

```python
test_list = [['forward', '5'], ['down', '5'], ['forward', '8'], ['up', '3'], ['down', '8'],['forward', '2']]
```

```python
depth , horizontal_position, aim  = 0, 0, 0

for sub_list in data_radar_list_split:
    depth, horizontal_position, aim = update_position(sub_list)

print('depth:',depth, '\n','horizontal_postion:' ,horizontal_position, '\n','aim:',aim)

answer = depth * horizontal_position
print('answer:',answer)
```

```python
depth , horizontal_position, aim  = 0, 0, 0

for sub_list in test_list:
    depth, horizontal_position, aim = update_position(sub_list)

print('depth:',depth, '\n','horizontal_postion:' ,horizontal_position, '\n','aim:',aim)

answer = depth * horizontal_position
print('answer:',answer)
```
