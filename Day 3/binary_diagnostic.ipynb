```python
#find the most common bit in a position
def find_most_common_bit(list, positon_bit):

    counter_positon_bit = 0
    
    for i in range(len(list)):
        if list[i][positon_bit] == '1':
            counter_positon_bit += 1
    
    average_bit_pos = counter_positon_bit/len(list)
    
    if average_bit_pos > 0.5:
        return '1'
    else:
        return '0'
```

```python
#find the most common bit for all postions
def find_most_common_bit_all_pos(list):
    list_bits = []
    for i in range(len(list[0])):
        list_bits.append(find_most_common_bit(list, i))
    return list_bits
```

```python
input_bits_list = np.genfromtxt('binary_diagnostic.csv', delimiter='\n', dtype="U75")
input_bits_list = input_bits_list.tolist()
input_bits_list
```

```python
gamma = find_most_common_bit_all_pos(input_bits_list)
gamma_rate = ''.join(gamma)
gamma_rate
```

```python
#create the opposite binary list of gamma
gamma_opposite = ['0' if x == '1' else '1' for x in gamma]
gamma_opposite = ''.join(gamma_opposite)
epsilon_rate = gamma_opposite
```

```python
#transform the binary numbers to decimal
epsilon_rate_number = int(epsilon_rate, 2)
gamma_rate_number = int(gamma_rate, 2)

power_consumption = gamma_rate_number * epsilon_rate_number
power_consumption
```

```python
def gamma_rate_tier(list):
    gamma_rate_tier = find_most_common_bit_all_pos(list)
    gamma_rate_tier = ''.join(gamma_rate_tier)
    return gamma_rate_tier
```

```python
def create_mask_gamma(list, position, gamma_value):
    list = np.array(list)
    mask = [True if list[i][position] == gamma_value[position] else False for i in range(len(list))]
    return mask
```

```python
def apply_mask_gamma(list, position):
    list = np.array(list)
    gamma_value = gamma_rate_tier(list)
    list2 = list[create_mask_gamma(list, position, gamma_value)].copy()
    return list2
```

```python
def apply_mask_all_pos(list):
    listA = np.array(list)
    for i in range(len(list[0])):
        listA = apply_mask_gamma(listA, i)
    return listA
```

```python
def find_least_common_bit(list, positon_bit):
    
        counter_positon_bit = 0
        
        for i in range(len(list)):
            if list[i][positon_bit] == '1':
                counter_positon_bit += 1
        
        average_bit_pos = counter_positon_bit/len(list)
        
        if average_bit_pos > 0.5:
            return '0'
        else:
            return '1'
```

```python
def find_least_common_bit_all_pos(list):
    list_bits = []
    for i in range(len(list[0])):
        list_bits.append(find_least_common_bit(list, i))
    return list_bits
```

```python
def epsilon_rate_tier(list):
    epsilon_rate_tier = find_least_common_bit_all_pos(list)
    epsilon_rate_tier = ''.join(epsilon_rate_tier)
    return epsilon_rate_tier
```

```python
def create_mask_epsilon(list, position, epsilon_value):
    list = np.array(list)
    mask = [True if list[i][position] == epsilon_value[position] else False for i in range(len(list))]
    return mask
```

```python
def apply_mask_epsilon(list, position):
    list = np.array(list)
    epsilon_value = epsilon_rate_tier(list)
    list2 = list[create_mask_epsilon(list, position, epsilon_value)].copy()
    return list2
```

```python
def apply_mask_all_pos_epsilon(list):
    listA = np.array(list)
    for i in range(len(list[0])):
        listA = apply_mask_epsilon(listA, i)
        if len(listA) == 1:
            break
    return listA
```
