```python
import numpy as np
```

```python
raw_data_thermal = np.genfromtxt('input_thermal.txt', delimiter='\n', dtype="U75" )
raw_data_thermal
```

```python
#split data
data_thermal = raw_data_thermal.tolist().copy()
data_thermal = [x.split('->') for x in data_thermal]

#remove empty spaces
data_thermal = [[x.strip() for x in y] for y in data_thermal] 

#split positions
data_thermal = [[x.split(',') for x in y] for y in data_thermal]

#transform strings to floats
data_thermal = [[[int(x) for x in y] for y in z] for z in data_thermal]

data_thermal
```

```python
#create a matrix 9x9
map_matrix = np.zeros((1000,1000), dtype=int)
map_matrix
```

```python
#draw lines between points
def draw_line(x1,y1,x2,y2):
    if x1 == x2:
        for i in range(min(y1,y2),max(y1,y2)+1):
            map_matrix[x1,i] += 1
    elif y1 == y2:
        for i in range(min(x1,x2),max(x1,x2)+1):
            map_matrix[i,y1] += 1
    else:
        slope = (y2-y1)/(x2-x1)
        if slope > 0:
            for i in range(min(x1,x2),max(x1,x2)+1):
                map_matrix[i,int(y1+slope*(i-x1))] += 1
        else:
            for i in range(min(x1,x2),max(x1,x2)+1):
                map_matrix[i,int(y1+slope*(i-x1))] += 1



for x in data_thermal:
    draw_line(x[0][0],x[0][1],x[1][0],x[1][1])

        
map_matrix
```

```python
print(np.count_nonzero(map_matrix>1))
```
